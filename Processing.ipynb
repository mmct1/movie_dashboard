```python
df = pd.read_csv('./data/movie_ratings.csv')
df.head()
df= df.dropna()

df = df[df['genres'] != 'unknown']
```

```python
df.to_csv('./data/movie_data_reshaped.csv',index=False)
```

```python
# What's the breakdown of genres for the movies that were rated?
df_genres= df['genres'].value_counts().to_frame().reset_index()
df_genres.to_csv('./data/genre_data.csv',index=False)

dftest=pd.read_csv('./data/genre_data.csv')
dftest

```

```python
# 2. Which genres have the highest viewer satisfaction (highest ratings)?
df.groupby('genres')['rating'].mean().reset_index()
```

```python
# 3. How does mean rating change across movie release years?
df.groupby('year')['rating'].agg(['mean','median']).reset_index()
```
